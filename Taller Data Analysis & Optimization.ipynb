### <center><font color = 'lightblue'><strong>Modelo de Clasificación</strong></font></center>

Para el desarrollo de un modelo de clasificación utilizamos un modelo basado en máquinas de soporte vectorial clásico, este modelo nos permite encontrar una regla de clasificación sobre un subconjunto de datos desde otro subconjunto dada una serie de características.

El modelo toma su base en programación cuadrática y tiene la siguiente forma,

$$ \min\limits_{\beta, \beta_0} \frac{1}{2} || \beta ||^{2} + C \sum_{i=1}^{N} \epsilon_i $$

Sujeto a un conjunto de restricciones

$$ y_i(x_i^T \beta + \beta_0) \geq 1 - \epsilon_i, \hspace{1cm} \forall i \in \{1, \dots, N \} $$
$$ \epsilon_i \geq 0, \hspace{4cm} \forall i \in \{1, \dots, N \} $$

En lo anterior $|| \beta  || $ representa el módulo vectorial de la forma,

$$ || \beta ||^2 = \sqrt{(\beta_1^2 + \beta_2^2 + \dots + \beta_p^2)} = \sqrt{\sum_{i=1}^{p} \beta_i^2} $$

Donde $p$ corresponde al número de columnas en el conjunto de datos, y $N$ el tamaño de la muestra

#### Generación de conjuntos de entrenamiento y prueba


Para llevar a cabo el proceso de ajustar el modelo separamos los datos en dos conjuntos, uno de entrenamiento que sirve para obtener los valores de las variables $\beta$  y $\beta_0$ , y otro conjunto para evaluar el desempeño del modelo

```python
import numpy as np

iris_df = iris_df.sample(frac = 1)
X = iris_df.drop('Label', axis = 1)
y = iris_df['Label']
X = X.to_numpy()
y = y.to_numpy()
y = np.where(y == 0, -1, 1)
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)

y_test
```

```python
pip install gurobipy
```

```python
from gurobipy import *
```

Primero tenemos que definir el modelo en nuestro ambiente

```python
quadratic_model = Model('quadratic')
```

Añadimos las variables de decisión del modelo

```python
u = {}
c = {}
```

```python
for i in range(X_train.shape[1]):
  u[i] = quadratic_model.addVar(vtype = GRB.CONTINUOUS, name = 'u' + str(i))
```

```python
u
```

```python
for i in range(len(X_train)):
  c[i] = quadratic_model.addVar(vtype = GRB.CONTINUOUS, name = 'c' + str(i))

bi = quadratic_model.addVar(vtype = GRB.CONTINUOUS, name = 'bi')

quadratic_model.update()
```

```python
quadratic_model
```

Definimos la función objetivo

```python
obj_fn = 0.5 * (sum(u[i]**2 for i in range(X_train.shape[1]))) + 1.0 * (sum(c[i] for i in range(len(X_train))))

quadratic_model.setObjective(obj_fn, GRB.MINIMIZE)
quadratic_model.update()
```

```python
quadratic_model
```

Añadimos las restricciones del modelo de programación cuadrática

```python
for i in range(len(X_train)):
  quadratic_model.addConstr(y_train[i] * (sum(u[j] * X_train[i][j] for j in range(X_train.shape[1])) - bi) - 1 + c[i] >= 0)
  quadratic_model.addConstr(c[i] >= 0)

quadratic_model.update()

quadratic_model
```

```python
quadratic_model.setParam('OutputFlag', True)
quadratic_model.optimize()
```

```python
def find_letter(letter, lst):
    return any(letter in word for word in lst)
```

```python
for v in quadratic_model.getVars():
    if find_letter('c', v.varName) == False:
        print('%s: %g' % (v.varName, v.x))
```

```python
predict_gurobi = []

for i in range(len(X_test)):
    predict_gurobi.append(sum(u[j].x*X_test[i][j] for j in range(X_train.shape[1])) - bi.x)

predict_gurobi = np.sign(predict_gurobi)

resultados = pd.DataFrame({'Actual': y_test, 'Gurobi': predict_gurobi})
resultados['Gurobi'] = (resultados['Gurobi'].astype(int))
```

```python
resultados
```

Los resultados del modelo anterior muestran los pesos de las variables $\beta$ y $\beta_0$  para cada una de las componentes del conjunto de datos, resulta interesante que esto podamos plotearlo a través de funciones lineales (ya que el modelo que utilizamos es lineal), así tenemos $\beta_1$  para el SepalLength, $\beta_2$  para el SepalWidth, $\beta_3$  para el PetalLength, y $\beta_4$ para el PetalWidth, lo que nos permite construir rectas que separan los conjuntos, así (lo siguiente considerando solo dos componentes)

$$ \beta_4×\text{Petal Width}+\beta_3×\text{Petal Length}+\beta_0=0, $$

$$ \text{Petal Width}=−\frac{\beta_3}{\beta_4}×\text{Petal Length}−\frac{\beta_0}{\beta_4} $$

lo que sigue siendo una función lineal de la forma  $y=mx+b$

### <center><font color = 'lightblue'><strong>Evaluación de la clasificación</strong></font></center>

De la matriz de confusión asociada al problema de clasificación construímos lo siguientes indicadores que permiten evaluar el desempeño del modelo y qué tan bien clasifica la regla generada, estos indicadores son,

$$ \text{Accuracy} = \frac{TN + TP}{TN + FN + TP + FP} $$

$$ \text{Sensitivity} = \frac{TP}{TP + FN} $$

$$ \text{Especificidad} = \frac{TN}{TN + FP} $$

Todas estas métricas pueden ser construídas a través de módulos pre-built en python mediante scikit-learn

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(resultados['Actual'], resultados['Gurobi']))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

cm = confusion_matrix(resultados['Actual'], resultados['Gurobi'])

print('Accuracy score igual a: ' + str(accuracy_score(resultados['Actual'], resultados['Gurobi'])))
print('Sensitivity igual a: ' + str(precision_score(resultados['Actual'], resultados['Gurobi'])))
print('Especificidad igual a: ' + str(recall_score(resultados['Actual'], resultados['Gurobi'])))
```
